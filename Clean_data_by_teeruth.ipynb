```python
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords


# Define a function to tokenize and remove stopwords
def remove_stopwords(text):
    stopwords = list(thai_stopwords())
    list_word = word_tokenize(text)
    list_word_not_stopwords = [i for i in list_word if i not in stopwords]
    return ' '.join(list_word_not_stopwords)

# Tokenize and remove stopwords for all rows in the 'story' column
df['story_cleaned'] = df['story'].apply(remove_stopwords)

# Print the first few rows to verify the result
print(df.head())
```
